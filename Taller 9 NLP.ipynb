# Taller 9 - Nube de palabras

```python
import pandas as pd
```

```python
data = pd.read_csv(r'D:/NLP/Clase11-31-10-2020/Princesas.csv')
data.head()
```

```python
import numpy as np

#nube de palabras
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
 
from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

#librerias de imagen
import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
#hacer nube de palabras corpus (lo que la gente dice en el corpus)
palabras = data.Personalidad.str.cat(sep=" ")
wordcloud = WordCloud().generate(palabras)
 
plt.imshow(wordcloud)
plt.show()
```

```python
# Más parametros de nube de palabras en https://www.datacamp.com/community/tutorials/wordcloud-python
palabras = data.Personalidad.str.cat(sep=' ')
 
wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='PuRd',
                     stopwords=stopwords).generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
 
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
#cargar imagen
princesa = PIL.Image.open("D:/NLP/Clase11-31-10-2020/princesa.jpg")
display(princesa)
```

```python
#crear mascara : pasar imagen a números

mask=np.array(princesa)
mask
```

```python
palabras = data.Personalidad.str.cat(sep=' ')
 
wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='spring',
                     stopwords=stopwords,
                     mask=mask, contour_width=0.5, contour_color='pink').generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
 
wordcloud.to_file("princesa.jpg")
 
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
