```python
import re
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

# Punto 1: Pre-Procesamiento

- `[11 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[11 pts]` Crear una nueva columna con el texto en min√∫scula, sin caracteres especiales ni n√∫meros y sin palabras vacias

```python
#Leer el archivo 1

dialogos=pd.read_csv(r'C:\Users\toshiba\Documents\CURSO NLP\dialogos.csv', sep=',')
dialogos
```

Crear una nueva columna con el texto en min√∫scula, sin caracteres especiales ni n√∫meros y sin palabras vacias

```python
#crear nueva columna y convertir a min√∫sculas 
dialogos['pre-procesado'] = dialogos['Locuci√≥n'].str.lower() # Min√∫scula
dialogos
```

```python
#Conservar s√≥lo palabras y Tokenizar

dialogos['pre-procesado']= dialogos['pre-procesado'].apply(lambda x: re.sub(r"[\W\d_]+", " ", str(x)))# Dejar s√≥lo letras

#dialogos['pre-procesado']=dialogos['pre-procesado'].str.strip() #Quitar espacios del principio

dialogos
```

```python
#quitar palabras vacias

dialogos['pre-procesado'] = dialogos['pre-procesado'].apply(lambda x: ' '.join([palabra for palabra in x.split() if palabra not in (stopwords_sp)]))
dialogos

```

# Punto 2: Representaci√≥n vectorial

- `[11 pts]` Crear una bolsa de palabras (BoW) del corpus usando la columna pre-procesada
- `[11 pts]` ¬øCu√°ntas palabras hay en el vocabulario? (Usando la funci√≥n de `sklearn`)

```python
from sklearn.feature_extraction.text import CountVectorizer


count_vect = CountVectorizer()
bow_rep = count_vect.fit_transform(dialogos['pre-procesado'].values)#transformar el texto

 

#print(f"Vocabulario: {count_vect.vocabulary_}")

 

print("Representaci√≥n BoW de 'dialogos': ", bow_rep[0].toarray())
```
